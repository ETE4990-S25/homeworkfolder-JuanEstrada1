```python
import os
import hashlib

def get_checksum(file_path):
    hash_obj = hashlib.md5()
    try:
        with open(file_path, 'rb') as f:
            while chunk := f.read(8192):
                hash_obj.update(chunk)
        return hash_obj.hexdigest()
    except (IOError, PermissionError):
        return None

def find_duplicates(directory):
    if not os.path.isdir(directory):
        print("Invalid directory. Please try again.")
        return
    files_by_name = {}
    for root, _, files in os.walk(directory):
        for file in files:
            files_by_name.setdefault(file, []).append(os.path.join(root, file))
    
    files_by_checksum = {}
    for file_name, paths in files_by_name.items():
        if len(paths) > 1: 
            for path in paths:
                checksum = get_checksum(path)
                if checksum:
                    files_by_checksum.setdefault(checksum, []).append(path)
    
    found = False
    for checksum, paths in files_by_checksum.items():
        if len(paths) > 1:
            found = True
            print(f"\nDuplicate files (MD5: {checksum}):")
            for path in paths:
                print(f"  {path}")
    if not found:
        print("No duplicates found.")

def menu():
    while True:
        print("File Duplicate Finder")
        print("1. Enter directory to search")
        print("2. Exit")
        choice = input("Choose an option: ")
        if choice == '1':
            directory = input("Enter directory path: ")
            print(f"Searching for duplicates in {directory}...")
            find_duplicates(directory)
        elif choice == '2':
            print("Exiting")
            break
        else:
            print("Invalid choice.")

if __name__ == "__main__":
    menu()
```
